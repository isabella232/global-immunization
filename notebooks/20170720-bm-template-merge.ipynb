### Functions

```python
def counter(item):
    return 1

def filterCountries(ctype):
    if (ctype == 'Sovereign country'):
        return True
    elif (ctype == 'Country'):
        return True
    elif (ctype == 'Lease'):
        return True
    else:
        return False
```

### Imports

```python
df_countryNames = pd.read_csv('../data/imports/country-names.csv')
```

### Analysis

```python
# Group to create count and apply counter function
df_countryNamesGrouped = df_countryNames.reset_index()
df_countryNamesGrouped['index'] = df_countryNamesGrouped['index'].apply(counter)
df_countryNamesGrouped.groupby('TYPE').sum().reset_index()
# df_countryNamesGrouped
```

```python
list(df_countryNames)
```

```python
# Filter for only a few categories
df_countryFiltered = df_countryNames
df_countryFiltered['Filter'] = df_countryFiltered['TYPE'].apply(filterCountries)
df_countryFiltered = df_countryFiltered[df_countryFiltered['Filter'] == True]
df_countryFiltered = df_countryFiltered.reset_index().drop(['Filter', 'index'], axis=1)
df_countryFiltered
```

```python
# List columns to keep
list(df_countryNames)
cols = ['ISO_A3','ADM0_DIF','LEVEL','ADM0_A3','ADM0_A3_IS','GU_A3','ISO_N3','TYPE','GEOUNIT','ABBREV','NAME','NAME_LONG']

# Filter entire country list
df_condCountryNames = df_countryNames[cols]
df_condCountryNames

# Filter soverign country-only list
df_condFilteredCountryNames = df_countryFiltered[cols]
df_condFilteredCountryNames

# df_condCountryNames[df_condCountryNames['ADM0_A3_US'] == 'MDV']
# list(df_countryNames)
```

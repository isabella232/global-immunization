```python
# -*- coding: utf-8 -*-
%matplotlib inline

import pandas as pd
import numpy as np
import math
import matplotlib as mplstyle
import matplotlib.pyplot as plt
import locale
import arrow
from locale import atof
from dateutil import parser
from datetime import datetime
```

### Functions

```python
def percentChange(val1, val2):
    val1 = round(val1, 2)
    val2 = round(val2, 2)
    percChange = (val2-val1)/val1
    return round((percChange * 100),2)
```

### Import

```python
df = pd.read_csv('../data/imports/dtp1.csv')
```

### By Country

```python
# Drop vaccine column
df_clean = df.drop('vaccine', axis=1)
```

```python
# Find percent change
df_clean['percChange'] = np.vectorize(percentChange)(df_clean['1996'],df_clean['2016'])
```

```python
# df_clean.sort_values('percChange')
df_under90 = df_clean[df_clean['2016'] < 90]
df_under90.sort_values('unicef_region')
```

```python
df_under90.to_csv('../data/exports/dtp1-df_under90.csv')
```

### By Region

```python
# Group by region
df_byRegion = df_clean.groupby('unicef_region').mean().reset_index().sort_values('2016',ascending=False)
```

```python
# Find percent change
df_byRegion['percChange'] = np.vectorize(percentChange)(df_byRegion['1986'],df_byRegion['2016'])
```

```python
df_byRegion.sort_values('2016')
```
